```python
import tkinter as tk
from tkinter import messagebox

def Fahrenheit(c):
    c = float(c)
    f = (c*(9/5)) + 32
    return f

def Celsius(f1):
    f1= float(f1)
    c1 = (f1-32)/(9/5)
    return c1

def ctof():
    c = entry_c.get()
    result = Fahrenheit(c)
    label_result.config(text =f"Fahrenheit Degree = {result}")

def ftoc():
    f1 = entry_f1.get()
    result1 = Celsius(f1)
    label1_result.config(text = f"Celcius Degree = {result1}")


def reset_fields():
 entry_c.delete(0, "end")

def reset_fields1():
 entry_f1.delete(0, "end")
 

root = tk.Tk()
root.title("Temperature Converter")
root.geometry("500x400")

label_c = tk.Label(root, text = "Celsius")
label_c.grid(row = 0,column=0)

label_f1 = tk.Label(root, text = "Fahrenheit")
label_f1.grid(row =4, column = 0)


entry_c = tk.Entry(root)
entry_c.grid(row = 0, column = 1)

entry_f1 = tk.Entry(root)
entry_f1.grid(row = 4, column = 1)

button = tk.Button(root, text = "Convert",command = ctof)
button.grid(row = 1, column = 1)

button1 = tk.Button(root,text = "Convert", command = ftoc)
button1.grid(row = 5, column = 1)

reset_button= tk.Button(root, text="Reset", command=reset_fields)
reset_button.grid(row=0,column=2)

reset_button1 = tk.Button(root, text = "Reset", command = reset_fields1)
reset_button1.grid(row = 4, column = 2)

label_result = tk.Label(root)
label_result.grid(row = 2, column = 1, columnspan=2)

label1_result = tk.Label(root)
label1_result.grid(row = 6, column = 1, columnspan = 2)

root.mainloop()
```
